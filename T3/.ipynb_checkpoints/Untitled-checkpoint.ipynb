```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Read the dataset
data = pd.read_csv("CAR_PRICE_T3.csv")

```

```python
# Data Cleaning

# Check for missing values
print(data.isnull().sum())
```

```python
# Drop rows with missing values
data.dropna(inplace=True)
```

```python
# Visualizations and EDA

# Explore the distribution of the target variable
plt.figure(figsize=(8, 6))
sns.histplot(data['price'], kde=True)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Distribution of Car Prices')
plt.show()
```

```python
# Explore the correlation between numerical features and the target variable
corr_matrix = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Exploratory Data Analysis (EDA)
# Correlation heatmap
corr = data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Distribution of the target variable
plt.figure(figsize=(8, 6))
sns.histplot(data['price'], kde=True)
plt.title('Distribution of Price')
plt.show()
```

```python
# Scatter plots for numerical variables
num_cols = ['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginesize',
            'boreratio', 'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg']

plt.figure(figsize=(12, 18))
for i, col in enumerate(num_cols):
    plt.subplot(5, 3, i+1)
    plt.scatter(data[col], data['price'])
    plt.xlabel(col)
    plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
# Box plots for categorical variables
cat_cols = ['fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel', 'enginelocation',
            'enginetype', 'cylindernumber', 'fuelsystem']

plt.figure(figsize=(12, 18))
for i, col in enumerate(cat_cols):
    plt.subplot(5, 2, i+1)
    sns.boxplot(x=data[col], y=data['price'])
    plt.xlabel(col)
    plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

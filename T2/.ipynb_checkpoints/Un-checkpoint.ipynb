```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load the dataset
data = pd.read_csv('Unemployment in India.csv')
```

```python
# Handle missing values
imputer = SimpleImputer(strategy='mean')
data[[' Estimated Employed', ' Estimated Unemployment Rate (%)']] = imputer.fit_transform(data[[' Estimated Employed', ' Estimated Unemployment Rate (%)']])
```

```python
# Check if the expected column exists in the dataset
target_column = ' Estimated Unemployment Rate (%)'
if target_column not in data.columns:
    raise KeyError(f"Column '{target_column}' not found in the dataset.")
```

```python
# Data cleaning
data.dropna(inplace=True)
```

```python
# Data visualization
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y=' Estimated Unemployment Rate (%)', data=data)
plt.title("Unemployment Rate by Region")
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize the relationships between features and the target variable
plt.figure(figsize=(10, 8))
sns.pairplot(data, x_vars=[' Estimated Employed',  ' Estimated Labour Participation Rate (%)'], y_vars=' Estimated Unemployment Rate (%)', kind='scatter')
plt.show()
```

```python
# Additional visualizations
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y=' Estimated Employed', data=data)
plt.title("Employment by Region")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y=' Estimated Labour Participation Rate (%)', data=data)
plt.title("Labour Participation Rate by Region")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data[' Estimated Unemployment Rate (%)'], kde=True)
plt.title(" Distribution of Unemployment Rate")
plt.xlabel(" Unemployment Rate (%)")
plt.show()
```

```python
# Feature selection
X = data[[' Estimated Employed', ' Estimated Labour Participation Rate (%)']]
y = data[' Estimated Unemployment Rate (%)']
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=120)
```

```python
# Train the models
models = [
    LinearRegression(),
    RandomForestRegressor(),
    KNeighborsRegressor(),
    DecisionTreeRegressor()
]

for model in models:
    model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print(f"Model: {model.__class__.__name__}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R^2 Score: {r2:.2f}")
print()
```

```python
# Plotting the predicted values against the actual values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Unemployment Rate (%)")
plt.ylabel("Predicted Unemployment Rate (%)")
plt.title(f"{model.__class__.__name__} - Actual vs Predicted")
plt.show()
```
